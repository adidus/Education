```python
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt
%matplotlib inline
```

# Log-normal distribution
https://en.wikipedia.org/wiki/Log-normal_distribution

```python
def plot_lognorm(n, subsets):
    mu, sigma = 1., 0.7
    s = np.random.lognormal(mu, sigma, subsets)
    count, bins, ignored = plt.hist(s, n, normed=True, align='mid', label='LOGNORM')
    x = np.linspace(min(bins), max(bins), 10000)
    pdf = (np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2)) / (x * sigma * np.sqrt(2 * np.pi)))
    plt.plot(x, pdf, linewidth=3, color='y', label='PDF')
    plt.axis('tight')
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
    plt.legend(loc='upper right')
    plt.show()
    
```

For Log-Normal distribution with mu = 1 and sigma = 0.7 our mean = exp(1+(0.7^2)/2) = 3.47 and var = exp(2*1+0.7^2)(exp(0.7^2)âˆ’1) = 7.62

```python
plot_lognorm(100, 1000)
```

```python
plot_lognorm(200, 1000)
```

```python
plot_lognorm(500, 1000)
```

```python
In conclusion. Log-normal distribution is a continuous probability distribution of a random variable whose logarithm is normally distributed. 
From our graphics we can understand that CLT is right and can be applied to LOG-Normal distribution.
```
